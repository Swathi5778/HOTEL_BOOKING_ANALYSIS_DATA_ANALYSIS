**6-Which country do most travelers come from?**

**List the 5 countries with the most passengers.**

```python
count_country=df.groupby(by='country')['name'].count()
count_country=count_country.sort_values(ascending=False)
df_count_country=pd.DataFrame(count_country)

print("The most travelers come from:\n ",count_country[:1])

print("\nList the 5 countries with the most passengers.\n",count_country[:5])
```

```python
import matplotlib.pyplot as plt
# Plotting the bar graph
plt.figure(figsize=(10, 6))
count_country[:5].plot(kind='bar', color='skyblue')
plt.title('Top 5 Countries with the Most Travelers')
plt.xlabel('Country')
plt.ylabel('Number of Travelers')
plt.xticks(rotation=45)
plt.show()
```

```python
import matplotlib.pyplot as plt
# Plotting the pie chart
plt.figure(figsize=(10, 6))
count_country[:5].plot(kind='pie', autopct='%1.1f%%', colors=['skyblue', 'lightgreen', 'lightcoral', 'lightskyblue', 'lightpink'], startangle=140)
plt.title('Top 5 Countries with the Most Travelers')
plt.ylabel('')  # Hide the y-label as it's not needed in a pie chart
plt.show()
```

**7-Who has the most ADR? How much?**

```python
ADR=df.groupby(by='name')['adr'].max()
ADR=ADR.sort_values(ascending=False)
ADR_df = pd.DataFrame(ADR)
print('The most ADR(Average Daily Rate) is:\n',ADR_df.iloc[0])
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
# Plotting the bar plot using Seaborn
plt.figure(figsize=(12, 6))
sns.barplot(x=ADR.index[:10], y=ADR.values[:10], palette='viridis')
plt.title('Top 10 Names with the Highest Average Daily Rate (ADR)')
plt.xlabel('Name')
plt.ylabel('Average Daily Rate (ADR)')
plt.xticks(rotation=45)
plt.show()
```

**8-Specify the mean of the total ADR?** (two decimal places)

```python
num_adr=df['adr'].count()
Avg_adr=df['adr'].sum()/num_adr
print('Average of total Average Daily Rate is:', Avg_adr.round(decimals=2))
```

```python
import matplotlib.pyplot as plt
adr_counts = df['adr'].value_counts()
# Plotting the donut chart
fig, ax = plt.subplots()
ax.pie(adr_counts, labels=adr_counts.index, autopct='%1.1f%%', startangle=90)
centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
ax.axis('equal')  
plt.title('Distribution of ADR')
plt.show()
```

**9-Specify the average number of nights spent.** (two decimal places)

```python
df['all_night']= df.apply(lambda x: x['stays_in_weekend_nights'] + x['stays_in_week_nights'], axis=1)
m = df['all_night'].mean()
np.round(m,2)
```

```python
import matplotlib.pyplot as plt

# Calculate the counts of unique total nights stayed values
all_night_counts = df['all_night'].value_counts()

# Plotting the pie chart
plt.figure(figsize=(8, 8))
plt.pie(all_night_counts, labels=all_night_counts.index, autopct='%1.1f%%', startangle=90)

# Equal aspect ratio ensures that pie is drawn as a circle.
plt.axis('equal')
plt.title('Distribution of Total Nights Stayed')
plt.show()
```

**12-Who has booked the hotel with the most children and babies?**

```python
max_babies=df['babies'].idxmax()
max_children=df[ 'children'].idxmax()

print(df.iloc[max_children]["name"],
      " has booked a hotel with the largest number of children.")

print(df.iloc[max_babies]["name"],
      " has booked a hotel with the largest number of babies.")
```

```python
import matplotlib.pyplot as plt

# Find the index of the row with the largest number of children and babies
max_children_idx = df['children'].idxmax()
max_babies_idx = df['babies'].idxmax()

# Get the names and counts
max_children_name = df.loc[max_children_idx, 'name']
max_babies_name = df.loc[max_babies_idx, 'name']
max_children_count = df.loc[max_children_idx, 'children']
max_babies_count = df.loc[max_babies_idx, 'babies']

# Create bar graph
plt.bar(['Max Children', 'Max Babies'], [max_children_count, max_babies_count], color=['blue', 'orange'])
plt.title('Hotels with Largest Number of Children and Babies')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```
